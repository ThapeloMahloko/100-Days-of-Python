## Python Comments

Comments are used to make the code more readable and to explain the code logic to anyone who reads it. Comments are not executed by the Python interpreter, so they have no impact on the program's output or performance. They are solely for the benefit of humans reading the code.

There are two main types of comments in Python:

1. **Single-line Comments**:
   - Single-line comments start with a hash symbol (`#`). Everything following the `#` on that line is considered a comment and is ignored by the interpreter.
   - Example:
     ```python
     # This is a single-line comment
     print("Hello, World!")  # This comment is ignored by Python
     ```

2. **Multi-line Comments**:
   - Python does not have a specific syntax for multi-line comments, but you can create them by using a series of single-line comments or by using triple quotes (`'''` or `"""`). While triple quotes are typically used for docstrings (documentation strings), they can also serve as multi-line comments if not assigned to a variable or used in a function.
   - Example using single-line comments:
     ```python
     # This is a multi-line comment
     # written over multiple
     # lines
     ```
   - Example using triple quotes:
     ```python
     """
     This is a multi-line comment.
     It can span multiple lines.
     As many lines as you want.
     """
     print("Hello, World!")
     ```

**When to Use Comments:**
- To explain complex sections of code.
- To describe the purpose of a function or a block of code.
- To temporarily disable parts of your code for debugging purposes.

**Good Practices:**
- Keep comments concise and to the point.
- Use comments to explain *why* something is done, not *what* is being done (the code itself should be clear enough to explain the "what").
- Avoid redundant comments that state the obvious, such as `# Increment x by 1` right before `x += 1`.

By effectively using comments, you can make your Python code easier to understand and maintain.

Variables are fundamental concepts in programming that allow you to store and manipulate data. In Python, variables are created when you assign a value to them. You do not need to declare the type of a variable in advance because Python is dynamically typed.

### Creating Variables

To create a variable, you simply assign a value to it using the equals sign (`=`):

```python

```

#### Examples:

1. **Integer Variable**:

3. **String Variable**:

#### Variable Naming Rules:
- Variable names must start with a letter (a-z, A-Z) or an underscore (_).
- The rest of the name can contain letters, numbers (0-9), and underscores.
- Variable names are case-sensitive (`name`, `Name`, and `NAME` are three different variables).
- Avoid using Python reserved keywords like `class`, `for`, `while`, etc.

#### Examples of Valid Variable Names:
```python
user_age = 30
total_cost = 99.99
is_valid = False
_name = "John"
```

#### Examples of Invalid Variable Names:
```python
2nd_place = "Silver"  # Cannot start with a number
total-cost = 100.0    # Hyphens are not allowed
for = 10              # 'for' is a reserved keyword
```

#### Assigning Multiple Variables:

### When to Use Integers, Floats, Strings, and Booleans

Choosing the appropriate data type is essential for writing efficient and readable code. Here’s a brief guide on when to use each of these common data types in Python:

#### **Integers (`int`)**
- **Definition**: Whole numbers without a fractional component.
- **Use When**:
  - Counting discrete items (e.g., number of students, number of cars).
  - Performing arithmetic operations where fractions are not required.
  - Using loop counters or indexing elements in sequences.
  - Representing simple statuses (e.g., 0 for off, 1 for on).

  **Example**:
  ```python
  age = 25
  num_apples = 10
  ```

#### **Floats (`float`)**
- **Definition**: Numbers with a fractional component, represented by a decimal point.
- **Use When**:
  - Dealing with measurements that require precision (e.g., height, weight).
  - Performing arithmetic operations that include fractions or require precision.
  - Representing probabilities or percentages.
  - Working with scientific calculations or data that involve decimal points.

  **Example**:
  ```python
  height = 5.9
  price = 19.99
  temperature = 98.6
  ```

#### **Strings (`str`)**
- **Definition**: A sequence of characters enclosed in single, double, or triple quotes.
- **Use When**:
  - Storing and manipulating text data (e.g., names, addresses, messages).
  - Representing data that is naturally textual, like file names or web addresses.
  - Displaying messages or prompts to users.
  - Reading and writing text data from and to files.

  **Example**:
  ```python
  name = "Alice"
  address = "123 Main St."
  greeting = "Hello, World!"
  ```

#### **Booleans (`bool`)**
- **Definition**: A data type with only two possible values: `True` or `False`.
- **Use When**:
  - Making decisions using conditional statements (e.g., `if`, `while`).
  - Representing binary states (e.g., on/off, yes/no, true/false).
  - Storing the results of logical operations or comparisons.
  - Controlling the flow of a program based on certain conditions.

  **Example**:
  ```python
  is_student = True
  has_paid = False
  is_logged_in = True
  ```
  
#### Example Usage in a Program:
Here’s a practical example combining all these data types:

````python
name = "Alice"               # String
age = 30                     # Integer
height = 5.4                 # Float
is_student = True            # Boolean
final_cost = 245.00 * 0.65


print("Name:", name)                         # String
print("Age:", age)                           # Integer
print("Height:", height)                     # Float
print("Is student:", is_student)             # Boolean
print("Final cost after discount: R", final_cost)  # Float
````

In this example:
- We use a string to store the name.
- We use an integer to store the age.
- We use floats to handle the height, discount, and cost calculations.
- We use a boolean to determine if the person is a student and apply a discount accordingly.

Understanding when to use each data type will help you write more efficient, clear, and maintainable Python code.

### Basic Arithmetic

In Python, basic arithmetic operations include addition, subtraction, multiplication, and division. Here's a brief explanation of each:

1. **Addition (`+`):**

````python
a = 3
b = 4

print(a + b)
````

   - Example: `10 / 2` results in `5.0`.

````python
a = 10
b = 3

print(a / b)
````

5. **Integer Division (`//`):**
   - The double-slash operator performs integer division, returning the quotient as an integer.

These basic arithmetic operations are fundamental in programming and are commonly used for various calculations in Python and other programming languages.

#### Popular Equations in Python


Import Math package

````python
import cmath
import math
````

1. **Quadratic Equation:**
   - The quadratic equation is given by `ax^2 + bx + c = 0`, and the solutions can be found using the quadratic formula.

````python
a = 1
b = -3
c = 2

# Quadratic formula
solution1 = (-b + cmath.sqrt(b**2 - 4*a*c)) / (2*a)
solution2 = (-b - cmath.sqrt(b**2 - 4*a*c)) / (2*a)

print("Quadratic Solutions:", solution1, solution2)
````

2. **Pythagorean Theorem:**
   - The Pythagorean theorem states that in a right-angled triangle, the square of the hypotenuse is equal to the sum of the squares of the other two sides.

````python
side_a = 3
side_b = 4

# Pythagorean theorem
hypotenuse = math.sqrt(side_a**2 + side_b**2)

print("Hypotenuse length:", hypotenuse)
````

3. **Area of a Circle:**
   - The area of a circle is given by `πr^2`, where `r` is the radius.

````python
radius = 5

# Area of a circle
area_circle = math.pi * radius**2

print("Area of the circle:", area_circle)
````

4. **Factorial:**
   - The factorial of a non-negative integer `n` is the product of all positive integers less than or equal to `n`.

````python
n = 5

# Factorial
factorial_n = math.factorial(n)

print("Factorial of", n, ":", factorial_n)
````

#### Boolean Operations

Boolean operations in Python involve working with boolean values (`True` and `False`) and using logical operators to perform comparisons and combine boolean expressions. Here are some common boolean operations in Python:

1. **Comparison Operators:**
   - Comparison operators are used to compare values and return a boolean result.



````python
x = 5
y = 10

# Equal to
print(x == y)  # False

# Not equal to
print(x != y)  # True

# Greater than
print(x > y)   # False

# Less than
print(x < y)   # True

# Greater than or equal to
print(x >= y)  # False

# Less than or equal to
print(x <= y)  # True
````

2. **Logical Operators:**
   - Logical operators combine boolean values or expressions.



````python
a = True
b = False

# Logical AND
print(a and b)  # False

# Logical OR
print(a or b)   # True

# Logical NOT
print(not a)    # False
````

3. **Membership Operators:**
   - Membership operators test if a value is a member of a sequence (e.g., a string, list, or tuple).



````python
my_list = [1, 2, 3]

# In
print(2 in my_list)  # True

# Not in
print(4 not in my_list)  # True
````

4. **Identity Operators:**
   - Identity operators compare the memory address of two objects.



````python
x = [1, 2, 3]
y = [1, 2, 3]

# is
print(x is y)   # False

# is not
print(x is not y)  # True
````

These boolean operations are fundamental for making decisions, controlling flow, and performing conditional logic in Python programs. They are extensively used in `if` statements, loops, and other control structures.

# Arithmetic and compound assignment operators

### 1. Arithmetic Operators
These operators are used to perform basic mathematical operations between two variables.

- **Addition (`+`)**: Adds two numbers.
  ```python
  x + y  # Adds x and y
  ```
- **Subtraction (`-`)**: Subtracts the second number from the first.
  ```python
  x - y  # Subtracts y from x
  ```
- **Multiplication (`*`)**: Multiplies two numbers.
  ```python
  x * y  # Multiplies x by y
  ```
- **Division (`/`)**: Divides the first number by the second and returns a float.
  ```python
  x / y  # Divides x by y (returns a float)
  ```
- **Floor Division (`//`)**: Divides the first number by the second, removing the decimal part (returns an integer).
  ```python
  x // y  # Divides x by y and returns an integer
  ```
- **Modulus (`%`)**: Returns the remainder when dividing the first number by the second.
  ```python
  x % y  # Gives the remainder of x divided by y
  ```
- **Exponentiation (`**`)**: Raises the first number to the power of the second number.
  ```python
  x ** y  # Raises x to the power of y
  ```

### 2. Compound Assignment Operators
These combine an arithmetic operation with assignment, allowing you to modify the value of a variable in place.

- **Add and Assign (`+=`)**: Adds and then assigns the result to the variable.
  ```python
  x += y  # x = x + y
  ```
- **Subtract and Assign (`-=`)**: Subtracts and then assigns the result to the variable.
  ```python
  x -= y  # x = x - y
  ```
- **Multiply and Assign (`*=`)**: Multiplies and then assigns the result to the variable.
  ```python
  x *= y  # x = x * y
  ```
- **Divide and Assign (`/=`)**: Divides and then assigns the result to the variable (returns a float).
  ```python
  x /= y  # x = x / y
  ```
- **Floor Divide and Assign (`//=`)**: Floor divides and then assigns the result to the variable (returns an integer).
  ```python
  x //= y  # x = x // y
  ```
- **Modulus and Assign (`%=`)**: Calculates the remainder and assigns it to the variable.
  ```python
  x %= y  # x = x % y
  ```
- **Exponentiate and Assign (`**=`)**: Raises the variable to a power and then assigns the result.
  ```python
  x **= y  # x = x ** y
  ```

````python
x = 5
y = 10

# Addition
print(x + y)  # 15  (Adds x and y)

# Subtraction
print(x - y)  # -5  (Subtracts y from x)

# Multiplication
print(x * y)  # 50  (Multiplies x by y)

# Division
print(y / x)  # 2.0  (Divides y by x, returns a float)

# Floor division
print(y // x)  # 2  (Divides y by x, returns an integer by removing the fractional part)

# Modulus
print(y % x)  # 0  (Returns the remainder of the division of y by x)

# Exponentiation
print(x ** 2)  # 25  (Raises x to the power of 2)

# Compound assignment operators

# Add and assign
x += y  # x = x + y -> x becomes 15
print(x)  # 15

# Subtract and assign
x -= 5  # x = x - 5 -> x becomes 10
print(x)  # 10

# Multiply and assign
x *= 2  # x = x * 2 -> x becomes 20
print(x)  # 20

# Divide and assign
x /= 5  # x = x / 5 -> x becomes 4.0
print(x)  # 4.0

# Floor divide and assign
x //= 3  # x = x // 3 -> x becomes 1.0
print(x)  # 1.0

# Modulus and assign
x %= 2  # x = x % 2 -> x becomes 1.0
print(x)  # 1.0

# Exponentiate and assign
x **= 3  # x = x ** 3 -> x becomes 1.0
print(x)  # 1.0
````
